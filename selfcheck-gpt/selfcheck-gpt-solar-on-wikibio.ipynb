```python
from openai import OpenAI
import os
import modeling_selfcheck_apiprompt

# Initialize SelfCheckAPIPrompt with the correct model
selfcheck_apiprompt = modeling_selfcheck_apiprompt.SelfCheckAPIPrompt(client_type="openai", 
                                                                      client=OpenAI(api_key=os.environ["UPSTAGE_API_KEY"], 
                                                                                    base_url="https://api.upstage.ai/v1/solar"), 
                                                                        model="solar-pro")
```

```python
sentences = ["Is this a test sentence?"]
sampled_passages = ["This is a context for testing."]
```

```python
scores = selfcheck_apiprompt.predict(sentences, sampled_passages, verbose=True)
print(scores)
```
