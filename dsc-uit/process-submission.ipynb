```python
import json
import numpy as np
from PIL import Image

def load_annotation(file_path):
    data = []

    with open (file_path, 'r', encoding="utf-8") as file:
        data = json.load(file)    
    return data

annotation_path = "data/vimmsd-public-test.json"
data = load_annotation(annotation_path)
```

```python
def map_label(label):
    return "multi-sarcasm" if label == 0 else "not-sarcasm"

export = []

for id in data:
    image_name = data[id]['image']
    predicted_label = map_label(result[int(id)]["predict"])

    export.append([id, image_name, predicted_label])
```

```python
import pandas as pd

export_df = pd.DataFrame(export, columns=["key", "image", "label"])
export_df.to_json('data/exports/combine_labels_ver0.1.json', orient='records')
```
