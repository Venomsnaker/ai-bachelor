```python
import json
import numpy as np
import pandas as pd
from PIL import Image

def load_annotation(file_path):
    data = []

    with open (file_path, 'r', encoding="utf-8") as file:
        data = json.load(file)    
    return data

annotation_path = "data/vimmsd-train.json"
public_test_path = "data/vimmsd-public-test.json"

data = load_annotation(annotation_path)
data_public_test = load_annotation(public_test_path)
```

```python
import pandas as pd

def map_image_to_type(label):
    if (label == "multi-sarcasm"):
        return 0
    else:
        return 1
    
result = []
for id in data:
    image_id = data[id]['image']
    label = map_image_to_type(data[id]['label'])
    text = data[id]['caption']
    image_id = image_id.replace(".jpg", "")
    result.append({"image_id": image_id, "text": text, "label": label})
    
public_test = []

for id in data_public_test:
    image_id = data_public_test[id]['image']
    label = data_public_test[id]['label']
    text = data_public_test[id]['caption']
    image_id = image_id.replace(".jpg", "")
    public_test.append({"image_id": image_id, "text": text, "label": 0})
```

```python
from sklearn.model_selection import train_test_split

df = pd.DataFrame(result)
train_df, test_val_df = train_test_split(df, test_size=0.3, random_state=42, stratify=df['label'])
val_df, test_df = train_test_split(test_val_df, test_size=0.4, random_state=42, stratify=test_val_df['label'])

train_result = train_df.to_dict(orient='records')
val_result = val_df.to_dict(orient='records')
test_result = test_df.to_dict(orient='records')

df_public_test = pd.DataFrame(public_test)
```

```python
print("Length train:", len(train_result))
print("Length valid:", len(val_result))
print("Length test:", len(test_result))
print("Length public testL", len(public_test))
```

```python
with open('data/datasets/train.json', 'w', encoding='utf-8') as f:
    json.dump(train_result, f, ensure_ascii=False, indent=4)
    
with open('data/datasets/valid.json', 'w', encoding='utf-8') as f:
    json.dump(val_result, f, ensure_ascii=False, indent=4)
    
with open('data/datasets/test.json', 'w', encoding='utf-8') as f:
    json.dump(test_result, f, ensure_ascii=False, indent=4)
    
with open('data/datasets/public_test.json', 'w', encoding='utf-8') as f:
    json.dump(public_test, f, ensure_ascii=False, indent=4)
```
