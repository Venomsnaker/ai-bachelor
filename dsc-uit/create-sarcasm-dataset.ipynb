```python
import json
import numpy as np
import pandas as pd
from PIL import Image

def load_annotation(file_path):
    data = []

    with open (file_path, 'r', encoding="utf-8") as file:
        data = json.load(file)    
    return data

annotation_path = "data/vimmsd-train.json"
data = load_annotation(annotation_path)
```

```python
import pandas as pd

def map_to_img_labels(label):
    if label in ["multi-sarcasm", "image-sarcasm"]:
        return "sarcasm"
    else:
        return "not-sarcasm"

result = []

for id in data:
    image_name = data[id]['image']
    label_test = map_to_img_labels(data[id]['label'])
    caption = data[id]['caption']
    
    result.append({'Image': image_name, 'Caption': caption, 'Label': label_test})
result = pd.DataFrame(result)
result.to_csv("data/datasets/train-sarcasm-dataset.csv", index=None)
```
